# Notebook de Limpieza

En este Notebook limpio el primer dataframe para quedarme con un listado de nombres de vehículos

```python
# Importo librerias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
import seaborn as sns
```

```python
# Importo el DF
df_extraccion = pd.read_csv('DF_extraccion.csv')
df_extraccion
```

```python
# Voy a limpiar la columna de nombres para comprobar cuantos vehículos únicos tengo
# Empiezo quitando los números de los nombres
df_EDA = df_extraccion
df_EDA['Nombre'] = df_EDA['Nombre'].str.replace(r'\d+\.', '', regex=True)
df_EDA['Nombre'] = df_EDA['Nombre'].str.lstrip() # Elimino los espacios en blanco que haya al principio de los nombres
df_EDA['Nombre'] = df_EDA['Nombre'].str.rstrip() # Elimino los espacios en blanco que haya al final de los nombres
df_EDA['Nombre'] = df_EDA['Nombre'].str.title() # Para normalizar los nombres y que sólo tengan en mayúscula la primera letra de cada palabra
# Existen diferentes marcas de coches que se pueden escribir de varias maneras, asi que las voy a normalizar reemplazando
df_EDA['Nombre'] = df_EDA['Nombre'].str.replace('Citroën', 'Citroen', regex=True)
df_EDA['Nombre'] = df_EDA['Nombre'].str.replace('Škoda', 'Skoda', regex= True)
df_EDA['Nombre'] = df_EDA['Nombre'].str.replace('Mercedes-Benz', 'Mercedes',  regex= True)

df_EDA
```

```python
# Miro los valores unicos en la columna Nombre
lista_nombres = df_EDA['Nombre'].unique().tolist()
lista_nombres
```
